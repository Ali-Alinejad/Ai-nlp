```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from torch.utils.data import DataLoader ,Dataset
from torch.optim import Adam
import torch.nn as nn
from torchsummary import summary
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
df = pd.read_csv("/content/rice-type-classification/riceClassification.csv")
df.head()
```

```python
print(df['Class'].value_counts())
```

```python
df.dropna(inplace = True)
df.drop(['id'],axis = 1,inplace = True)
print(df['Class'].value_counts())
```

```python
df.head()
```

```python
original_df = df.copy()
for column in df.columns:
  df[column] = df[column] / df[column].abs().max()
df.head()
```

```python
x = np.array(df.iloc[: ,:-1])
print(x)
```

```python
y = np.array(df.iloc[:,-1])
print(y)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.3)
x_test , x_val , y_test , y_val = train_test_split(x_test , y_test , test_size = 0.5)
```

```python
print(x_train.shape)
print(x_test.shape)
print(x_val.shape)
```

```python
class dataset(Dataset):
  def __init__(self,x,y):
    self.x = torch.tensor(x,dtype = torch.float32).to(device)
    self.y = torch.tensor(y,dtype = torch.float32).to(device)
  def __len__(self):
    return len(self.x)

  def __getitem__(self,index):
    return self.x[index] , self.y[index]

```

__init__: داده‌های x و y را به تنسور PyTorch تبدیل می‌کند و آن‌ها را روی GPU (یا CPU) قرار می‌دهد.

__len__: تعداد نمونه‌های دیتاست را برمی‌گرداند.

__getitem__: هنگام دریافت یک نمونه خاص، ویژگی‌ها (x[index]) و لیبل (y[index]) را برمی‌گرداند.

```python
trainig_data = dataset(x_train , y_train)
validation_data = dataset(x_val ,y_val)
testing_data = dataset(x_test , y_test)
```

اینجا داری داده‌های خام رو تبدیل به کلاس dataset می‌کنی که قبلاً تعریف کردی. این کار باعث می‌شه داده‌ها به تنسور تبدیل بشن و روی دستگاه مناسب (CPU/GPU) ذخیره بشن.

توضیح خط به خط:
✅ dataset(x_train , y_train) → داده‌های آموزشی رو آماده می‌کنه.
✅ dataset(x_val , y_val) → داده‌های اعتبارسنجی رو آماده می‌کنه.
✅ dataset(x_test , y_test) → داده‌های تست رو آماده می‌کنه.

حالا برای استفاده در DataLoader آمادن.

```python
train_dataLoader = DataLoader(trainig_data, batch_size=8, shuffle=True)
val_dataLoader = DataLoader(validation_data, batch_size=8, shuffle=False)
test_dataLoader = DataLoader(testing_data, batch_size=8, shuffle=False)
```

چرا DataLoader داده‌ها را دسته‌بندی (Batching) می‌کند؟
DataLoader در PyTorch داده‌ها را به Batchهای کوچک‌تر تقسیم می‌کند به چند دلیل:

1️⃣ بهینه‌سازی حافظه (Memory Efficiency)
اگر بخواهیم کل دیتاست را یکجا پردازش کنیم، حافظه (RAM/GPU) پر می‌شود.
✅ با پردازش داده‌ها در Batchهای کوچک، مصرف حافظه کاهش پیدا می‌کند.

2️⃣ افزایش سرعت پردازش (Parallelism)
مدل‌های PyTorch از GPU یا CPU استفاده می‌کنند و محاسبات ماتریسی را سریع انجام می‌دهند.
✅ با Batchهای کوچک، پردازش داده‌ها در موازی (Parallel) انجام می‌شود که باعث افزایش سرعت می‌شود.

3️⃣ بهبود عملکرد الگوریتم‌های بهینه‌سازی
الگوریتم‌های یادگیری مانند SGD بهتر روی Batchهای کوچک‌تر کار می‌کنند.
✅ باعث می‌شود گرادیان‌ها پایدارتر باشند و مدل بهتر یاد بگیرد.

4️⃣ جلوگیری از نوسانات زیاد (Gradient Stability)
اگر هر بار یک نمونه به مدل بدهیم، گرادیان‌ها خیلی تغییر می‌کنند و یادگیری نوسانی می‌شود.
✅ با Batch، گرادیان‌ها هموارتر می‌شوند و یادگیری پایدارتر است.

```python
hidden_neurons = 10

class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()

        self.input_layer = nn.Linear(x.shape[1], hidden_neurons)
        self.hidden_layer = nn.Linear(hidden_neurons, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.input_layer(x)
        x = self.hidden_layer(x)
        x = self.sigmoid(x)
        return x

model = MyModel().to(device)
```

این مدل یک شبکه عصبی ساده با یک لایه ورودی و یک لایه مخفی است. هدف این مدل، پیش‌بینی یک خروجی بین ۰ و ۱ (برای مشکلات دسته‌بندی دودویی) است.

__init__: در این قسمت، لایه‌ها و توابع فعال‌سازی مدل تعریف می‌شوند.

input_layer: یک لایه خطی است که تعداد ویژگی‌های ورودی (x.shape[1]) را به تعداد نورون‌های مخفی (hidden_neurons) نگاشت می‌کند.

hidden_layer: لایه‌ای است که نورون‌های مخفی را به یک نورون خروجی متصل می‌کند.

sigmoid: تابع فعال‌سازی سیگموید برای محدود کردن خروجی بین ۰ و ۱، مناسب برای دسته‌بندی دودویی.

forward: این تابع نحوه عبور داده‌ها از مدل را تعریف می‌کند.

داده‌های ورودی ابتدا از لایه ورودی عبور می‌کنند.

سپس از لایه مخفی عبور کرده و در نهایت خروجی به وسیله تابع سیگموید تولید می‌شود.

```python
summary(model,(x.shape[1],))
```

کد summary(model, (x.shape[1],)) از کتابخانه torchsummary استفاده می‌کند تا جزئیات ساختار مدل را به نمایش بگذارد. این جزئیات شامل ابعاد ورودی‌ها، تعداد پارامترها، نوع لایه‌ها و غیره می‌شود. در واقع، این تابع به شما کمک می‌کند تا بتوانید به سرعت معماری مدل خود را مشاهده کنید و بفهمید که هر لایه مدل چه تعداد پارامتر دارد.

در اینجا:

model: مدل شما که قبلاً تعریف کرده‌اید (در اینجا MyModel).

(x.shape[1],): این پارامتر به تابع summary ابعاد ورودی به مدل را می‌دهد. به طور خاص، x.shape[1] تعداد ویژگی‌های ورودی را نشان می‌دهد.

```python
criterion = nn.BCELoss()
optimizer = Adam(model.parameters(),lr = 1e-3)
```


1. criterion = nn.BCELoss()
این خط از کد معیار خطا را برای مدل تعیین می‌کند. در اینجا از Binary Cross-Entropy Loss (BCELoss) استفاده می‌شود که معمولاً برای مسائل دسته‌بندی دودویی (binary classification) به کار می‌رود.

BCELoss: این معیار خطا برای پیش‌بینی‌هایی که به صورت احتمال (بین 0 و 1) هستند و هدف مدل هم پیش‌بینی یکی از دو کلاس (مثلاً 0 یا 1) است، مناسب است. این تابع میزان انحراف پیش‌بینی مدل از مقدار واقعی را محاسبه می‌کند.

برای مثال، اگر مدل شما احتمال "1" را پیش‌بینی کند و برچسب واقعی "0" باشد، BCELoss یک مقدار خطا بزرگ تولید می‌کند و برعکس.

2. optimizer = Adam(model.parameters(), lr=1e-3)
این خط از کد بهینه‌ساز مدل را تعیین می‌کند. در اینجا از Adam استفاده شده است که یک روش بهینه‌سازی پرکاربرد در یادگیری عمیق است.

Adam: یک الگوریتم بهینه‌سازی است که ترکیبی از دو روش بهینه‌سازی Momentum و RMSprop است. این روش برای بهینه‌سازی وزن‌های مدل استفاده می‌شود. مزیت اصلی آن در این است که به طور خودکار نرخ یادگیری (learning rate) را تنظیم می‌کند و به مدل کمک می‌کند که سریعتر و دقیقتر به سمت کمترین مقدار خطا (minima) حرکت کند.

model.parameters(): این قسمت تمام پارامترهای قابل آموزش مدل را به بهینه‌ساز می‌دهد. این پارامترها همان وزن‌ها و بایاس‌هایی هستند که در طول فرآیند آموزش باید به روز شوند.

lr=1e-3: این مقدار، نرخ یادگیری (learning rate) بهینه‌ساز را تعیین می‌کند که سرعت تغییرات وزن‌های مدل را مشخص می‌کند. مقدار 1e-3 معادل 0.001 است و معمولاً برای شروع، نرخ یادگیری مناسبی است.

خلاصه:
BCELoss برای محاسبه خطا در مسائل دسته‌بندی دودویی استفاده می‌شود.

Adam به عنوان یک بهینه‌ساز برای به روز رسانی وزن‌های مدل استفاده می‌شود و نرخ یادگیری آن را 1e-3 (0.001) تنظیم می‌کند.

```python
total_loss_train_plot = []
total_loss_validation_plot = []
total_acc_train_plot = []
total_acc_validation_plot = []

epochs = 10

for epoch in range(epochs):
    total_acc_train = 0
    total_loss_train = 0
    total_acc_val = 0
    total_loss_val = 0

    # Training loop
    for data in train_dataLoader:
        inputs, labels = data
        prediction = model(inputs).squeeze(1)
        batch_loss = criterion(prediction, labels)
        total_loss_train += batch_loss.item()
        acc = ((prediction).round() == labels).sum().item()
        total_acc_train += acc
        batch_loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    # Validation loop
    with torch.no_grad():
        for data in val_dataLoader:
            inputs, labels = data
            prediction = model(inputs).squeeze(1)
            batch_loss = criterion(prediction, labels)
            total_loss_val += batch_loss.item()
            acc = ((prediction).round() == labels).sum().item()
            total_acc_val += acc

    # Average loss and accuracy calculation
    avg_train_loss = total_loss_train / len(train_dataLoader)
    avg_train_acc = total_acc_train / len(train_dataLoader.dataset)
    avg_val_loss = total_loss_val / len(val_dataLoader)
    avg_val_acc = total_acc_val / len(val_dataLoader.dataset)

    # Storing the results
    total_loss_train_plot.append(avg_train_loss)
    total_loss_validation_plot.append(avg_val_loss)
    total_acc_train_plot.append(avg_train_acc)
    total_acc_validation_plot.append(avg_val_acc)

    # Print the results
    print(f'Epoch [{epoch+1}/{epochs}], '
          f'Train Loss: {avg_train_loss:.4f}, Train Accuracy: {avg_train_acc:.4f}, '
          f'Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {avg_val_acc:.4f}')
```

کد شما آموزش مدل را در چند مرحله انجام می‌دهد:

آماده‌سازی متغیرها: متغیرها برای ذخیره‌سازی دقت و خطای مجموع در هر epoch تعریف شده‌اند.

حلقه epoch: برای هر epoch، مدل آموزش می‌بیند و خطا و دقت محاسبه می‌شود.

حلقه batch: داده‌ها از train_dataLoader گرفته می‌شوند و مدل پیش‌بینی می‌کند. خطا (loss) محاسبه و دقت به‌دست می‌آید. سپس این مقادیر به‌طور تجمعی در total_loss_train و total_acc_train ذخیره می‌شوند.

break: در حال حاضر کد فقط یک batch را پردازش می‌کند، زیرا دستور break حلقه‌ها را متوقف می‌کند. برای آموزش کامل باید breakها حذف شوند.

هدف این کد آموزش مدل و ثبت مقادیر خطا و دقت برای هر epoch است.
